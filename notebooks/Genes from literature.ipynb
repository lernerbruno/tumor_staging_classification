```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/cleaned.csv')
df.head()
```

```python
df_literature_genes = pd.read_csv('data/literature_genes.csv')

good_codes = []
for code in df_literature_genes['Code']:
    good_code = df.columns[df.columns.str.startswith(code)].values[0]
    good_codes.append(good_code)
len(set(good_codes))
```

```python
other_genes = set(df.columns[df.columns.str.startswith('ENSG')]) - set(good_codes)
others_to_add = np.random.choice(list(other_genes), size=7, replace=False)
```

```python
lit_genes = df[list(set(good_codes)) + list(others_to_add) + ["tumor_stage"]]
```

```python
lit_genes.head()
```

```python
lit_genes.to_csv('data/64_genes.csv')
```

## Data Preprocessing

```python
all_genes_df = lit_genes[lit_genes.columns[lit_genes.columns.str.startswith('ENSG')]]
```

```python
def normalize(c):
    return c/c.max()
all_genes_df = all_genes_df.apply(normalize, axis=0)
```

```python
all_genes_df
```

```python
final_df = pd.concat([all_genes_df, df['tumor_stage']], axis=1)
```

```python
final_df.head()
```

```python
final_df.to_csv('data/literature_genes_df_normalized.csv')
```
