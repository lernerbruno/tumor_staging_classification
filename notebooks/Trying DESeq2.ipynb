```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/64_genes.csv').drop(columns=['Unnamed: 0'])
df.head()
```

```python
genes = df.drop(columns=['tumor_stage'])
genes = genes.T
```

```python
genes['id'] = genes.index
```

```python
sample_df = pd.DataFrame({'samplename': genes.columns}) \
        .query('samplename != "id"')
sample_df.index = sample_df.samplename
sample_df
```

```python
from diffexpr.py_deseq import py_DESeq2

dds = py_DESeq2(count_matrix = genes,
               design_matrix = sample_df,
               design_formula = '~ sample',
               gene_column = 'id') # <- telling DESeq2 this should be the gene ID column
    
dds.run_deseq() 
dds.get_deseq_result()
res = dds.deseq_result 
res.head()
```
